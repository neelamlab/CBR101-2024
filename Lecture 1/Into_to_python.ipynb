### Vectors

In Numpy, a 1D array is often referred to as a vector. Vectors are fundamental mathematical entities that represent a sequence of numbers. Numpy provides a powerful array object (numpy.ndarray) to work with vectors efficiently.

```python
vector = np.array([1, 2, 3, 4, 5])
print("1D Vector:", vector)

# Dot product of two vectors
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])

dot_product = np.dot(vector1, vector2)
print("Dot Product:", dot_product)
```

### Martix & Vectors

The equation `[A] * [x] = [y]` represents a matrix-vector multiplication, where:

*    [A] is a matrix.
*    [x] is a column vector.
*    [y] is the resulting column vector.

\
\
Key Points:

* The number of columns in [A] must be equal to the number of rows in [x] for the multiplication to be valid.
* The resulting vector [y] will have the same number of rows as [A] and the same number of columns as [x].

```python
import numpy as np

# matrix [A]
matrix_A = np.array([[2, 3, -1],
                    [4, 1, 2],
                    [3, 2, -2]])
# column vector [x]
vector_x = np.array([[1],
                     [2],
                     [3]])

# matrix-vector multiplication
vector_y = np.dot(matrix_A, vector_x)

print("Matrix [A]:\n", matrix_A)
print("Vector [x]:\n", vector_x)
print("Resulting Vector [y]:\n", vector_y)
```

```python
# Calculate the inverse of A
try:
    inverse_A = np.linalg.inv(matrix_A)
    print("Inverse of Matrix A:\n", inverse_A)
except np.linalg.LinAlgError:
    print("Matrix A is singular, and its inverse does not exist.")
```

```python
# Calculate the inverse of A
try:
    inverse_A = np.linalg.inv(matrix_A)

    # Solve for x using the equation x = A^-1 * y
    vector_x = np.dot(inverse_A, vector_y)
    print("Solution for x:\n", vector_x)
except np.linalg.LinAlgError:
    print("Matrix A is singular, and its inverse does not exist.")
```

The pseudoinverse of a matrix AA, denoted as A+A+, satisfies the condition \( A * A^+ * A = A \)

```python
A_pseudoinverse = np.linalg.pinv(matrix_A)

result = np.dot(np.dot(matrix_A, A_pseudoinverse), matrix_A)

print("Original Matrix A:\n", matrix_A)
print("Pseudoinverse of Matrix A:\n", A_pseudoinverse)
print("Result of A * A^+ * A:\n", result)
```

## Noise

```python
# Set a seed for reproducibility
np.random.seed(42)

# Generate a signal
t = np.linspace(0, 1, 1000)  # Time vector
signal = 5 * np.sin(2 * np.pi * 5 * t)  # A sine wave as the signal

# Generate random noise within different ranges
noise_range1 = np.random.uniform(-1, 1, len(t))  # Noise within [-1, 1]
noise_range2 = np.random.uniform(-10, 10, len(t))  # Noise within [-10, 10]

# Combine signal and noise
signal_with_noise1 = signal + noise_range1
signal_with_noise2 = signal + noise_range2
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(t, signal, label='Original Signal')
plt.ylim(-17,17)
plt.title('Original Signal')

plt.subplot(2, 2, 2)
plt.plot(t, noise_range1, label='Noise [-1, 1]')
plt.ylim(-17,17)
plt.title('Noise within [-1, 1]')

plt.subplot(2, 2, 3)
plt.plot(t, signal_with_noise1, label='Signal + Noise [-1, 1]')
plt.ylim(-17,17)
plt.title('Signal + Noise within [-1, 1]')

plt.subplot(2, 2, 4)
plt.plot(t, signal_with_noise2, label='Signal + Noise [-10, 10]')
plt.ylim(-17,17)
plt.title('Signal + Noise within [-10, 10]')

plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)

# Generate a signal
t = np.linspace(0, 1, 1000)  # Time vector
signal = 5 * np.sin(2 * np.pi * 5 * t)  # A sine wave as the signal

# Generate Gaussian noise
mean = 0
std_dev = 1
gaussian_noise = np.random.normal(mean, std_dev, len(t))
gaussian_noise_larger_variance = np.random.normal(mean, std_dev*3, len(t))

# Combine signal and Gaussian noise
signal_with_gaussian_noise = signal + gaussian_noise
signal_with_gaussian_noise_larger_variance = signal + gaussian_noise_larger_variance

# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(t, signal, label='Original Signal')
plt.title('Original Signal')

plt.subplot(3, 1, 2)
plt.plot(t, signal_with_gaussian_noise, label='Signal + Gaussian Noise')
plt.title('Signal + Gaussian Noise')

plt.subplot(3, 1, 3)
plt.plot(t, signal_with_gaussian_noise_larger_variance, label='Signal + High variance Gaussian Noise')
plt.title('Signal + Gaussian Noise')

plt.tight_layout()
plt.show()
```
